# CLUSTERING 

* Two forms of clustering 

* a. Flat Clustering 

* b. Hierarchical clustering

* With both forms of clustering, the machine is given a dataset, just the feature set and then the machine will search for 
  groups or cluster by itself and then each cluster is given an ID/label

* The difference between flat and hierarchical clustering is that in flat, you tell the machine the number of clusters you want
  it to find, while in hierarchical the machine figures out the group(clusters) and how many there ought to between


## 2. MEAN SHIFT (HIERARCHICAL CLUSTERING)

* Unlike Flat clustering, in Mean shift, the machine figures out how many clusters there ought to be and where those clusters actually are 

* In mean shift what we say is every datapoint is a cluster center(centroid).
* In mean shift we have the RADIUS and BANDWIDTH. Every datapoint has a bandwidth around it which is like a circle and the radius is the distance between that datapoint and the bandwidth. 
* After every single datapoint's bandwidth has been drawn, a datapoint will fall within the bandwidth of the other datapoints near it and other datapoints will also fall within it's own bandwidth.
* Then you take the mean of all the datapoints within that cluster(bandwidth), and with this new mean, comes a new bandwidth, then we take the mean of all the datapoints within our new bandwidth(bandwidth of the mean), then this has its own bandwidth also, and on and on...and eventually when that clusters center doesn't move anymore, it is optimized. You have convergence, and when you have convergence you are done.

## Importing all the necessary libraries

```python
import matplotlib.pyplot as plt
from matplotlib import style
import importlib
importlib.import_module('mpl_toolkits.mplot3d').Axes3D
style.use('ggplot')
import numpy as np
from sklearn.cluster import MeanShift
from sklearn.datasets.samples_generator import make_blobs
```

## Modelling 

```python
centers = [[1, 1, 1], [5, 5, 5], [3, 10, 10]]
X, _ = make_blobs(n_samples = 1000, centers = centers, cluster_std = 1) #Making random inputs using the make_blob function

ms = MeanShift()
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
print(cluster_centers)
n_clusters_ = len(np.unique(labels))
print('Number of estimated clusters:', n_clusters_)
```

## Visualization

```python
colors = 10*["r", "g","c","b","k"]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(X)):
    ax.scatter(X[i][0], X[i][1], X[i][2], c=colors[labels[i]], marker='o')

ax.scatter(cluster_centers[:,0], cluster_centers[:,1], cluster_centers[:,2],
           marker="x", color="k", s = 150, lw = 5, zorder = 10)        
plt.show()
```

### In the 3D plot above, we can see the 3 different clusters we have and their centroids
